```python
import numpy as np
from pynq import Xlnk
from pynq import Overlay
import time

ol = Overlay('bits/pynq_ad_fft/pynq_ad_fft.bit')
print("Load bits Done.")
print(dir(ol))
data_size = 512
```

```python
import time

d = ol.d2dma_0
while d.register_map.CTRL.AP_READY != 0:
    print("Waiting AP_READY...")
    time.sleep(1)
d.register_map.depth = data_size
d.register_map.CTRL.AP_START = 1
while d.register_map.CTRL.AP_DONE != 1:
    print("Waiting AP_DONE...")
    time.sleep(1)
print("ad2dma Done.")
```

```python
import pynq.lib.dma
dma = ol.axi_dma_0
xlnk = Xlnk()
input_buffer = xlnk.cma_array(shape=(16384//4,), dtype=np.uint32)
for i in range(data_size):
    input_buffer[i] = 0;
dma.recvchannel.transfer(input_buffer)
print("DMA receive done.")
if max(input_buffer) == 0:
    print("All Zero!!")
else:
    for i in range(data_size):
        print(hex(input_buffer[i]))
```
